# **プログラムの名称**

## **概要**

これは、整数が2～100項並んだ数列の一般項を求め、その先に予想される項を求めるプログラムです。

## **操作方法**

まず初めに**整数が2～100**項並んだ数列を**スペース区切り**で入力してください。また、ランダムで入力したい場合は"**/r**"と入力し、その後入力したい項数を**2～100の自然数**で入力してください。
次に、求めた一般項から予想される数列を、何項まで表示するかを**1000までの自然数**を入力してください。入力が完了すると計算が始まり、階差と一般項、計算で求められた値が出力されます。

また、出力される数字の桁数が大きいと、指数表記のfloat型からint型に変換する際に誤差が生じ、正しい結果が得られない場合があります。

## **制作者のコメント**

求めた一般項をΣを使わずに表示するために、式の分解を繰り返して法則性を見つけるところが一番苦労しました。また反省点は、指数表記のfloat型からint型に変換する際に生じる誤差の対応を仕切れなかったことです。

```python
#@title サブ処理
import string
import math
from IPython.display import Math
import random as r
import time

#丸め誤差対応、int変換
def arrange(x):
  #丸め誤差対応
  x = int(round(float(x),1))
  return x

#整数か調べる
def flost(s):
    if type(s) == list:
      tf = 0
      for o in s:
        try:
          float(o.replace('.','Hello, World!'))
        except ValueError:
          pass
        else:
          tf += 1
      if tf == len(s):
        return True
      else:
        return False
    else:
      try:
        float(s.replace('.','Hello, World!'))
      except ValueError:
        return False
      else:
        return True

#0,1表示非表示の分岐
def om0(x,l):
  if x == 0:
    return ""
  else:
    if l == 0:
      return x
    elif l == 1:
      return f"{x:+}"
    else:
      return f"{x}+"

def om0c(x,f,i):
  if f == 0 and i != 0:
    return x
  else:
    return f"({x})"

def om1(x):
  if x == 1:
    return ""
  elif x == -1:
    return "-"
  else:
    return x

#区切り線デザイン
def CL(x):
  st = 48
  if x == 0:
    return print("^^"*st)
  elif x == 1:
    return print("--"*st)
  elif x == 2:
    return print("=="*st)

#Σのアルファベット52以上対応
def ascii(x,i):
  f = i//52
  if i >= 52:
    return f"{string.ascii_letters[i-52*f]}_{f}"
  else:
    return string.ascii_letters[i]
```

```python
#@title メイン処理
delay = 0.2
#入力
print("💎<整数>が並んだ<数列>を2～100でスペース' '区切りで入力⌨️してください(例:>>1 3 8 100 2 -23 4096 -10)")
print("　乱数🎲で指定したい場合は'/r'と入力⌨️してください：")

A = ["Hello, World!"]
fi = 0
while len(A) <= 1 or len(A) > 100 or flost(A) == False:
  if A[0]=="/r":
    print("💎乱数🎲で指定したい<項数>を2～100で入力⌨️してください：")
    nram = "Hello, World!"
    ri = 0
    while not nram.isdecimal() or int(nram)<1 or int(nram)>100:
      if ri !=0:
        print('⚠️入力が適切ではありません、入力⌨️し直してください：')
      nram = str(input(">>"))
      ri = 1
    A = []
    for i in range(1,int(nram)+1):
      ram = str(r.randint(-256,1024))
      A.append(ram)
  else:
    if fi != 0:
      print('⚠️入力が適切ではありません、入力⌨️し直してください：')
    A = [str(e) for e in input(">>").split()]
  fi = 1



#与えられた数列とさらに予想される項の表示数入力
print("💎後に求まった一般項を使い、与えられた数列の先に予想される項を表示します")
print(f"　{len(A)}～1000でいくつ表示するか入力⌨️してください(例:>>{len(A)+40})：")

nl = "Hello, World!"
ki = 0
while not nl.isdecimal() or int(nl)<len(A) or int(nl)>1000:
  if ki !=0:
    print('⚠️入力が適切ではありません、入力⌨️し直してください：')
  nl = str(input(">>"))
  ki = 1

print('🔻入力⌨️に成功しました')

print("🚩与えられた数列の一般項の生成を開始します")
time.sleep(delay)
CL(2)

#A = A数列（与えられた数列）
#B = 数列保存（階差数列）
B = []
B.append([])

Br = [] #一時保存階差数列

Rb = [] #周回用階差数列
for b in A:
  B[0].append(arrange(b))
  Rb.append(arrange(b))
Rc = 0
d = 0


# 計算
for a in range(len(A)):
  for i in range(len(Rb)-1):
    Rc = Rb[i+1]-Rb[i]

    Br.append(arrange(Rc)) #階差数列を一時保存

  B.append(Br) #Bにリストを追加保存

  Rb = [] #今の数列リセット
  #次の数列のための今の階差数列保存
  for c in Br:
    Rb.append(c)

  Br = [] #一時保存した階差数列リセット

  #等差数列検出
  for k in range(len(B[a])-1):
    if B[a][k+1] == B[a][k]:
      d += 1
  if d == len(B[a])-1:
    if a == 0:
      time.sleep(delay)
      display(Math(fr"a_n={B[a][0]}")) #同じ数の数列だった場合
      break
    else:
      D = B[a][0] #公差
      time.sleep(delay)
      print(f"💡ここの等差数列の公差:{D}")

    #一番下の階差数列の一般項（等差数列）
    xM = f"{om1(D)}{string.ascii_letters[0]}{om0(arrange(B[a-1][0]-D),1)}"
    i = 0

    #与えられた数列の一般項（Σあり）
    while i != a-1:
      if i != a-2:
        xM = f"{om0(B[a-i-2][0],2)}\sum_{{{ascii(string.ascii_letters,i)}=1}}^{{{ascii(string.ascii_letters,i+1)}-1}}{om0c(xM,B[a-i-1][0],i)}"
      else:
        xM = f"{om0(B[a-i-2][0],2)}\sum_{{{ascii(string.ascii_letters,i)}=1}}^{{n-1}}{om0c(xM,B[a-i-1][0],i)}"
      i += 1

    time.sleep(delay)
    CL(1)
    print(f"🎓与えられた数列の一般項：")
    display(Math(f"a_n"))
    time.sleep(delay)
    if a != 1:
      display(Math(f"={xM}"))

    time.sleep(delay)
    #与えられた数列の一般項（Σなし）
    f = i
    m = 1
    yM = ""
    yMb = ""
    for fa in range(f):
      yM += f"(n-{1+fa})"
    if f == 0:
      yM += f"{om1(D)}n{om0(B[a-1][0]-D,1)}"
    elif f == 1:
      yM += fr"(\frac{{{om1(D)}n}}{{{math.factorial(f+1)}}}{{{om0(B[a-1][0]-D,1)}}})"
    else:
      yM += fr"(\frac{{{om1(D)}n}}{{{math.factorial(f+1)}}}+\frac{{{B[a-1][0]-D}}}{{{math.factorial(f)}}})"
    while f!=0:
      f -= 1
      m += 1
      yMb = ""
      if B[a-m][0] != 0:
        if f != 0:
          if B[a-m][0] == 1:
            yM += f"+"
          elif B[a-m][0] == -1:
            yM += f"-"
          else:
            yM += f"{B[a-m][0]:+}"
        else:
          yM += f"{B[a-m][0]:+}"
        for fb in range(f):
          yMb += f"(n-{1+fb})"
        if math.factorial(f) != 1:
          yM += fr"\frac{{{yMb}}}{{{math.factorial(f)}}}"
        else:
          yM += f"{yMb}"
    display(Math(fr"={yM}"))


    time.sleep(delay)
    CL(2)

    #与えられた数列とさらに予想される項
    for n in range(1,int(nl)+1):
      k = i
      w = 1

      wa = math.perm(n-1,k)*(D*n/math.factorial(k+1) + (B[a-1][0]-D)/math.factorial(k))
      while k!=0:
        k -= 1
        w += 1
        wa += B[a-w][0]*math.perm(n-1,k)/math.factorial(k)

      if len(nl) == 1:
        print(f"📝第{n:>1}項:{arrange(wa)}")
      elif len(nl) == 2:
        print(f"📝第{n:>2}項:{arrange(wa)}")
      elif len(nl) == 3:
        print(f"📝第{n:>3}項:{arrange(wa)}")
      else:
        print(f"📝第{n:>4}項:{arrange(wa)}")

      time.sleep(delay/10)
    break

  d = 0
  #階差数列表示
  if a == 0:
    print("⚙️与えられた数列")
  else:
    print(f"⚙️{a}段目階差数列")
  print(f"{B[a]}")
  time.sleep(delay)
  CL(0)
```
