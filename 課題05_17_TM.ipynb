```python
#@title ã‚µãƒ–å‡¦ç†
import string
import math
from IPython.display import Math
import random as r
import time

#ä¸¸ã‚èª¤å·®å¯¾å¿œã€intå¤‰æ›
def arrange(x):
  #ä¸¸ã‚èª¤å·®å¯¾å¿œ
  x = int(round(float(x),1))
  return x

#æ•´æ•°ã‹èª¿ã¹ã‚‹
def flost(s):
    if type(s) == list:
      tf = 0
      for o in s:
        try:
          float(o.replace('.','Hello, World!'))
        except ValueError:
          pass
        else:
          tf += 1
      if tf == len(s):
        return True
      else:
        return False
    else:
      try:
        float(s.replace('.','Hello, World!'))
      except ValueError:
        return False
      else:
        return True

#0,1è¡¨ç¤ºéè¡¨ç¤ºã®åˆ†å²
def om0(x,l):
  if x == 0:
    return ""
  else:
    if l == 0:
      return x
    elif l == 1:
      return f"{x:+}"
    else:
      return f"{x}+"

def om0c(x,f,i):
  if f == 0 and i != 0:
    return x
  else:
    return f"({x})"

def om1(x):
  if x == 1:
    return ""
  elif x == -1:
    return "-"
  else:
    return x

#åŒºåˆ‡ã‚Šç·šãƒ‡ã‚¶ã‚¤ãƒ³
def CL(x):
  st = 48
  if x == 0:
    return print("^^"*st)
  elif x == 1:
    return print("--"*st)
  elif x == 2:
    return print("=="*st)

#Î£ã®ã‚¢ãƒ«ãƒ•ã‚¡ãƒ™ãƒƒãƒˆ52ä»¥ä¸Šå¯¾å¿œ
def ascii(x,i):
  f = i//52
  if i >= 52:
    return f"{string.ascii_letters[i-52*f]}_{f}"
  else:
    return string.ascii_letters[i]
```
